```python
import pandas as pd

#Baca data dari file excel 
df_jabar = pd.read_excel('disperkim-od_16985_jumlah_produksi_sampah_berdasarkan_kabupatenkota_v3_data.xlsx', sheet_name='data')
df_jabar
```

```python
df_sampah = pd.DataFrame(df_jabar, columns=['nama_kabupaten_kota','jumlah_produksi_sampah','tahun'])
df_sampah

df_sampah.to_csv('data_sampah.csv')
df_sampah.to_excel('data_sampah.xlsx')
```

```python
#no.2
import numpy as np

d15 ={} #jumlah sampah pada tahun 2015

for i , row in df_sampah.iterrows():
    tahun = row['tahun']
    sampah = row['jumlah_produksi_sampah']

    if tahun == 2015 not in d15 :
        d15[tahun] = sampah
    elif tahun == 2015 in d15 :
        d15[tahun] += sampah
print  ('=============== Data Tahun 2015 ===============')
for i,j in d15.items():
    print(f"Jumlah sampah pada tahun {i} adalah {j:.2f} ton")
    
df_t2015 = pd.DataFrame ([
    {'Tahun':i,'Total/tahun':j}
    for i,j in d15.items()
])

df_t2015

df_t2015.to_csv('df_t2015.csv')
df_t2015.to_excel('df_t2015.xlsx')
```

```python
#NO.3

thn = {}

for i,j in df_sampah.iterrows():
     tahun = j['tahun']
     tsampah = j['jumlah_produksi_sampah']
     
     if tahun not in thn:
          thn[tahun] = tsampah 
     else :
          thn[tahun] += tsampah

print  ('=============== Data per tahun ==================')         
for i,j in thn.items():
     print(f'Pada tahun {i} terdapat {j:.2f} ton sampah')

df_sampahthn = pd.DataFrame([
   {'Tahun':i,'Total Sampah':j}
   for i,j in thn.items()
])

df_sampahthn

df_sampahthn.to_csv('df_sampahthn.csv')
df_sampahthn.to_excel('df_sampahthn.xlsx')
```

```python
#NO.4

tkk = {}

for i,j in df_sampah.iterrows():
     daerah = j['nama_kabupaten_kota']
     sampah = j['jumlah_produksi_sampah']
     
     if daerah not in tkk : 
          tkk[daerah] = sampah
     else : 
          tkk[daerah] += sampah
print  ('=============== Data per Daerah ===============')
for i,j in tkk.items():
     print(f'Pada daerah {i} terdapat {j:.2f} ton sampah')
     
df_sampahdaerah = pd.DataFrame([
     {'Nama Kabupaten/Kota':i,'Total Sampah':j}
     for i,j in tkk.items()
])
df_sampahdaerah

df_sampahdaerah.to_csv('df_sampahdaerah.csv')
df_sampahdaerah.to_excel('df_sampahdaerah.xlsx')
```
